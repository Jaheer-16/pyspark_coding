```python


data = [
    ("Alice", "Math", 85),
    ("Alice", "Science", 90),
    ("Alice", "History", 88),
    ("Bob", "Math", 70),
    ("Bob", "Science", 80),
    ("Bob", "History", 75),
    ("Charlie", "Math", 95),
    ("Charlie", "Science", 85),
    ("Charlie", "History", 92)
]
columns = ["student", "subject", "score"]
df = spark.createDataFrame(data, columns)

#df.show()



df2=df.groupBy('student').pivot('subject').agg(avg('score')).alias('avg_score')
df2.printSchema()

df2.show()
```

```python
from pyspark.sql.functions import sum


sales = [
    ("StoreA", "Jan", 100),
    ("StoreA", "Feb", 150),
    ("StoreA", "Mar", 200),
    ("StoreB", "Jan", 80),
    ("StoreB", "Feb", 120),
    ("StoreB", "Mar", 160),
    ("StoreC", "Jan", 90),
    ("StoreC", "Feb", 110),
    ("StoreC", "Mar", 130)
]
columns = ["store", "month", "sales"]
df_sales = spark.createDataFrame(sales, columns)


df2=df_sales.groupBy('month').pivot('store').agg(sum('sales')).alias('total_sales')
df2.show()

```
