```python
from pyspark.sql.functions import*

data = [
    (1, "Alice", 85, 90),
    (2, "Bob", 70, 80),
    (3, "Charlie", 95, 85)
]
columns = ["id", "name", "math_score", "science_score"]
df = spark.createDataFrame(data, columns)

df2=df.withColumn('score',array('math_score','science_score'))

df2.show()
```

```python
data = [
    ("Jaheer", ["btech", "phd"]),
    ("Alice", ["math", "science"]),
    ("Bob", None),          # null array
    ("Emma", [])            # empty array
]
columns = ["name", "education"]
df = spark.createDataFrame(data, columns)

df2=df.withColumn('flat_edu',explode_outer('education'))

df2.show()
```
