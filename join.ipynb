```python
employees = [
    (1, "John", "Sales"),
    (2, "Sarah", "Engineering"),
    (3, "Michael", "HR"),
    (4, "Anna", "Engineering"),
    (5, "David", "Sales")
]
emp_columns = ["emp_id", "name", "department"]
df_emp = spark.createDataFrame(employees, emp_columns)

departments = [
    ("Sales", "New York"),
    ("Engineering", "San Francisco"),
    ("HR", "Chicago"),
    ("Finance", "Boston")
]
dept_columns = ["department", "location"]
df_dept = spark.createDataFrame(departments, dept_columns)

# Now join safely

df_emp.join(df_dept, df_emp.department == df_dept.department, "inner").display()

#df_emp.join(df_dept, df_emp.department == df_dept.department, "right").display()
```

```python
employees = [
    (1, "John", "Sales"),
    (2, "Sarah", "Engineering"),
    (3, "Michael", "HR"),
    (4, "Anna", "Engineering"),
    (5, "David", "Sales"),
    (6, "Emma", "Finance")   # Finance not in departments
]
emp_columns = ["emp_id", "name", "department"]
df_emp = spark.createDataFrame(employees, emp_columns)


departments = [
    ("Sales", "New York"),
    ("Engineering", "San Francisco"),
    ("HR", "Chicago")
]
dept_columns = ["department", "location"]
df_dept = spark.createDataFrame(departments, dept_columns)



#df2=df_emp.join(df_dept, df_emp.department == df_dept.department, "left_semi")
df2=df_emp.join(df_dept, df_emp.department == df_dept.department, "left_anti")
df2.display()

#df_emp.join(df_dept, df_emp.department == df_dept.department, "right").display()
```

```python
students = [
    (1, "Alice"),
    (2, "Bob"),
    (3, "Charlie")
]
student_columns = ["student_id", "name"]
df_students = spark.createDataFrame(students, student_columns)


courses = [
    ("Math"),
    ("Science"),
    ("History")
]
course_columns = ["course"]
df_courses = spark.createDataFrame(courses, course_columns)


df_students.join(df_courses).display()
```

```python
data1 = [
    (1, "Alice", 20),
    (2, "Bob", 22)
]
columns1 = ["id", "name", "age"]
df1 = spark.createDataFrame(data1, columns1)


data2 = [
    ("Charlie", 3, 23, "Math"),
    ("David", 4, 25, "Science")
]
columns2 = ["name", "id", "age", "course"]
df2 = spark.createDataFrame(data2, columns2)



df1.display()
df2.display()

df3=df1.unionByName(df2,True)
df3.display()
```
