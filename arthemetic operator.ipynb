```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("ArithmeticDemo").getOrCreate()

data = [
    (1, "Alice", 10, 500),   # 10 items, each 500 price
    (2, "Bob", 5, 300),      # 5 items, each 300 price
    (3, "Charlie", 8, 450),  # 8 items, each 450 price
    (4, "David", 12, 700)    # 12 items, each 700 price
]

df = spark.createDataFrame(data, ["id", "name", "quantity", "unit_price"])
df.display()
```

```python
from pyspark.sql.functions import*

df1 = df.select('id','name','quantity',
    (col("unit_price")+100).alias("Increment_untprice"),

(col("unit_price")*100).alias("Multiply_untprice"),

(col("unit_price")/100).alias("Divide_untprice"),

(col("unit_price")-100).alias("Subtract_untprice")
)
df1.display()
```
